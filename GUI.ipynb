```python
# libraries
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
from tkinter import *


from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg, 
NavigationToolbar2Tk)
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np





#Covid Data#

df = pd.read_csv('covid.csv')
df1 = df.iloc[:,:26].copy()
df1 = df1.drop(columns=df1[["areaType","areaCode"]])
df1['date'] = pd.to_datetime(df1['date'])
df1.loc[df1['areaName'] == 'Oxfordshire']
df1['Total_Cases'] = df1.iloc[:,2:26].sum(axis=1).copy()
df1['Percentage_Change'] = df1.groupby(['areaName'])['Total_Cases'].pct_change()
df2 = df1.replace([np.inf, -np.inf], np.nan)
df2 = df2.fillna(0)
df3 = df2[df2['Total_Cases'] != 0]
df4 = df3.loc[df3['areaName'] == 'Oxfordshire']
df4.loc['date'] = pd.to_datetime(df4['date']).copy()

def graph1():
    func3 = df1.groupby(pd.Grouper(key='date', freq='1M'))['Total_Cases'].sum()
    func3.index = func3.index.strftime('%B')
    fig=plt.Figure(figsize =(6,5), dpi = 100)
    ax = fig.add_subplot(111)
    ax.set_title('Total number of Covid cases for All Areas')
           
    func3.plot(kind='line', color='red',legend=True, ax=ax,marker='*' );
        
    canvas = FigureCanvasTkAgg(fig, master = frame1)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1, padx=3, pady=3)
    
    toolbar = NavigationToolbar2Tk(canvas, frame1)
    toolbar.update()
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1,padx=3, pady=3)

def plot():
    window_1 = Toplevel(window)
    window_1.title("New Window")
    window_1.geometry("700x400")
    
    start_date = '2020-07-1'
    end_date = '2020-07-30'
    plot1= df4[(df4['date'] >= start_date) & (df4['date'] <= end_date)]
    func1 = (plot1[['date','Total_Cases']].set_index(["date"]))
    
    fig=plt.Figure(figsize =(8,7), dpi = 100)
    ax1 = fig.add_subplot(111)
    fig1_2=func1.plot(kind='bar',legend=True,color='red', ax=ax1)
    ax1.set_title('Total number of cases reported each day over a given date range in Oxfordshire:  start_date = 2020-07-1 end_date = 2020-07-30')
    
    canvas = FigureCanvasTkAgg(fig, master = window_1)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH, padx=3, pady=3 )
    
    toolbar = NavigationToolbar2Tk(canvas, window_1)
    toolbar.update()
    canvas.get_tk_widget().pack(side=tk.BOTTOM, fill=tk.BOTH, expand=1,padx=3, pady=3)
    
    
    plt.style.use('ggplot')
    plt.show()


def plot1():
    window1 = Toplevel(window)
    window1.title("New Window")
    window1.geometry("700x400")

  
    func2 = df4.groupby((df4["date"]).dt.isocalendar().week)['Total_Cases'].sum()
    fig=plt.Figure(figsize =(8,7), dpi = 100)
    ax2 = fig.add_subplot(111)
    fig2_1=func2.plot(kind='line',legend=True,color='red',marker='*', ax=ax2)
    ax2.set_title('Total number of cases reported each week over a given date range in Oxfordshire')
    
    canvas = FigureCanvasTkAgg(fig, master = window1)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    
    toolbar = NavigationToolbar2Tk(canvas, window1)
    toolbar.update()
    canvas.get_tk_widget().pack(side=tk.BOTTOM, fill=tk.BOTH, expand=1,padx=3, pady=3)
       
    plt.style.use('ggplot')
    plt.show()
    
       
    
    
    
    
def plot2():
    window2 = Toplevel(window)
    window2.title("New Window")
    window2.geometry("700x400")
    func3 = df4.groupby(pd.Grouper(key='date', freq='1M'))['Total_Cases'].sum()
    func3.index = func3.index.strftime('%B')
    
    fig=plt.Figure(figsize =(8,7), dpi = 100)
    ax3 = fig.add_subplot(111)
    func3.plot.area( color='red',legend=True, ax=ax3)
    ax3.set_title('Total number of cases reported each month over from the start of reporting to end in Oxfordshire')
    
    plt.style.use('ggplot')
    plt.show()
    
    canvas = FigureCanvasTkAgg(fig, master = window2)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    
    toolbar = NavigationToolbar2Tk(canvas, window2)
    toolbar.update()
    canvas.get_tk_widget().pack(side=tk.BOTTOM, fill=tk.BOTH, expand=1,padx=3, pady=3)
    
    
    
    
def plot3():
    window3 = Toplevel(window)
    window3.title("New Window")
    window3.geometry("700x400")
    
    start_date = '2020-07-01'
    end_date = '2020-07-30'
    df5 = df3[(df3['date'] >= start_date) & (df3['date'] <= end_date) & (df3['Total_Cases'] > 100 )]
    func4 = df5.groupby(df3['areaName'])['Total_Cases'].max().sort_values(ascending=True)
    
    fig=plt.Figure(figsize =(8,8), dpi = 100)
    ax4 = fig.add_subplot(111)
    ax4.set_title('Areas with the highest number of cases on a given day: start_date = 2020-07-01   end_date = 2020-07-30')
    
    func4.plot(kind='barh',legend=True,color='red', ax=ax4)
    plt.style.use('ggplot')
    plt.show()  
    
    canvas = FigureCanvasTkAgg(fig, master = window3)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH)
    
    toolbar = NavigationToolbar2Tk(canvas, window3)
    toolbar.update()
    canvas.get_tk_widget().pack(side=tk.BOTTOM, fill=tk.BOTH, expand=1,padx=3, pady=3)
    
    
def plot4():
    window4 = Toplevel(window)
    window4.title("New Window")
    window4.geometry("700x400")
    
    start_date = '2020-07-01'
    end_date = '2020-07-07'
    df6= df3[(df3['date'] >= start_date) & (df3['date'] <= end_date) & (df3['Percentage_Change'] > 2)]
    func5 = df6.groupby(df6['areaName'])['Percentage_Change'].max().sort_values(ascending=True)
    
    fig=plt.Figure(figsize =(8,8), dpi = 100)
    ax5 = fig.add_subplot(111)
    ax5.set_title('Areas with the largest positive change of cases for a given 7 day period: start_date =2020-07-01 end_date = 2020-07-07')
    
    func5.plot(kind='barh',legend=True, color='red', ax=ax5)
    plt.style.use('ggplot')
    plt.show()  
    
    canvas = FigureCanvasTkAgg(fig, master = window4)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH)
    
    toolbar = NavigationToolbar2Tk(canvas, window4)
    toolbar.update()
    canvas.get_tk_widget().pack(side=tk.BOTTOM, fill=tk.BOTH, expand=1,padx=3, pady=3)
    
    
def plot5():
    window5 = Toplevel(window)
    window5.title("New Window")
    window5.geometry("700x400")
    
    filter_list = ["Oxfordshire","Kent","Buckinghamshire","Berkshire","Hampshire",'Surrey','Sussex']
    df7 = df3[df3.areaName.isin(filter_list)]
    compare_dchange = pd.DataFrame(df7[["date","areaName","Percentage_Change"]])
    compare_dchange = compare_dchange.fillna(0)
    start_date = '2020-07-01'
    end_date = '2020-07-30'
    plot6= compare_dchange[(compare_dchange['date'] >= start_date) & (compare_dchange['date'] <= end_date)]
    plot6 = plot6.fillna(0)
    func6 = plot6.pivot_table(plot6,index ="date",columns = "areaName",aggfunc = np.mean)
    
    fig=plt.Figure(figsize =(8,8), dpi = 100)
    ax6 = fig.add_subplot(111)
    ax6.set_title('Comparison of two or more areas concerning their daily change over a given date range: start_date = 2020-07-01 end_date = 2020-07-30')
    
    func6.plot(kind='bar',legend=True,ax=ax6)
                          
    plt.style.use('ggplot')
    plt.show()  
    
    canvas = FigureCanvasTkAgg(fig, master = window5)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH)  
    
    toolbar = NavigationToolbar2Tk(canvas, window5)
    toolbar.update()
    canvas.get_tk_widget().pack(side=tk.BOTTOM, fill=tk.BOTH, expand=1,padx=3, pady=3)
    
def plot6():
    window6 = Toplevel(window)
    window6.title("New Window")
    window6.geometry("700x400")
    
    filter_list = ["Oxfordshire","Kent","Buckinghamshire","Berkshire","Hampshire",'Surrey','Sussex']
    df8 = df3[df3.areaName.isin(filter_list)]
    compare_cum_cases = pd.DataFrame(df8[["date","areaName","Total_Cases"]])
    start_date = '2020-07-01'
    end_date = '2020-07-30'
    plot7= compare_cum_cases[(compare_cum_cases['date'] >= start_date) & (compare_cum_cases['date'] <= end_date)]
    func7 = plot7.pivot_table(plot7,index ="date",columns = "areaName",aggfunc = np.mean)
    
    fig=plt.Figure(figsize =(8,8), dpi = 100)
    ax7= fig.add_subplot(111)
    ax7.set_title('Comparison of two or more areas concerning their cumulative cases total over a given date range: start_date = 2020-07-01  end_date = 2020-07-30')
        
    func7.plot(kind='bar',legend=True, ax=ax7)
    
    plt.style.use('ggplot')
    plt.show()
    
    canvas = FigureCanvasTkAgg(fig, master = window6)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH) 
    
    toolbar = NavigationToolbar2Tk(canvas, window6)
    toolbar.update()
    canvas.get_tk_widget().pack(side=tk.BOTTOM, fill=tk.BOTH, expand=1,padx=3, pady=3)
    

#STOP AND SEARCH

df_s_2=pd.read_csv('stop and search dataset/2.csv', sep=',')
df_s_3=pd.read_csv('stop and search dataset/3.csv', sep=',')
df_s_4=pd.read_csv('stop and search dataset/4.csv', sep=',')
df_s_5=pd.read_csv('stop and search dataset/5.csv', sep=',')
df_s_6=pd.read_csv('stop and search dataset/6.csv', sep=',')
df_s_7=pd.read_csv('stop and search dataset/7.csv', sep=',')
df_s_8=pd.read_csv('stop and search dataset/8.csv', sep=',')
df_s_9=pd.read_csv('stop and search dataset/9.csv', sep=',')
df_s_10=pd.read_csv('stop and search dataset/10.csv', sep=',')
df_s_11=pd.read_csv('stop and search dataset/11.csv', sep=',')
df_s_12=pd.read_csv('stop and search dataset/12.csv', sep=',')  

df_s = [df_s_2, df_s_3, df_s_4, df_s_5, df_s_6, df_s_7, df_s_8, df_s_9, df_s_10, df_s_11, df_s_12 ]
nan_value = 0

df1_s = pd.concat(df_s).fillna(nan_value).copy()
df1_s['Date'] = pd.to_datetime(df1_s['Date'], format='%Y-%m-%d')
df1_s = df1_s[df1_s['Age range'] != 0]
     
#Visulation All Age Range     
    
list1=['10-17', '18-24', '25-34','over 34','under 10' ]
def All_case(case):
    if case in list1:
        return 1
df1_s['All_Age_Case']=df1_s['Age range'].apply(All_case)

def graph1_s():
    func4_2 = df1_s.groupby(pd.Grouper(key='Date', freq='1M'))['All_Age_Case'].sum()
    func4_2.index = func4_2.index.strftime('%B')
    
    fig=plt.Figure(figsize =(6,5), dpi = 100)
    ax1s = fig.add_subplot(111)
    ax1s.set_title('All Age Stop & Search Tames Walley Police each month')
    func4_2.plot(kind='line',legend=True, color='red', ax=ax1s,marker='*')
    
    plt.style.use('ggplot')
    plt.show()
 
    canvas = FigureCanvasTkAgg(fig, master = frame1)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1,  padx=3, pady=3)

    toolbar = NavigationToolbar2Tk(canvas, frame1)
    toolbar.update()
    canvas.get_tk_widget().pack(side=tk.BOTTOM, fill=tk.BOTH, expand=1,padx=3, pady=3)
    
#Teen Age Each month
def Age(gender):
            
    if gender =='10-17' :
        return 1
    else:
        return 0
df1_s['teen']=df1_s['Age range'].apply(Age)

def plot2_s():
    window2s = Toplevel(window)
    window2s.title("New Window")
    window2s.geometry("700x400")
    
    func3_1 = df1_s.groupby(pd.Grouper(key='Date', freq='1M'))['teen'].sum()
    func3_1.index = func3_1.index.strftime('%B')
    
    fig=plt.Figure(figsize =(6,5), dpi = 100)
    ax2s = fig.add_subplot(111)
    ax2s.set_title('Teen Age Stop and Search Tames Walley Police each month')
    
    func3_1.plot(kind='line',legend=True, color='red', ax=ax2s, marker='*' )
    plt.style.use('ggplot')
    plt.show();
    
    canvas = FigureCanvasTkAgg(fig, master = window2s)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH) 
    
    toolbar = NavigationToolbar2Tk(canvas, window2s)
    toolbar.update()
    canvas.get_tk_widget().pack(side=tk.BOTTOM, fill=tk.BOTH, expand=1,padx=3, pady=3)

#Teen Age Each In July
def plot3_s():
    window3s = Toplevel(window)
    window3s.title("New Window")
    window3s.geometry("700x400")
    
    start_date = '2020-07-01'
    end_date = '2020-07-31'
    df3= df1_s[(df1_s['Date'] >= start_date) & (df1_s['Date'] <= end_date)]
    func2_1 = df3.groupby((df3["Date"]).dt.day)['teen'].sum()
    
    fig=plt.Figure(figsize =(6,5), dpi = 100)
    ax3s = fig.add_subplot(111)
    ax3s.set_title('Teen Age Stop and Search Tames Walley Police in July')
        
    func2_1.plot(kind='bar',legend=True, color='red', ax=ax3s)
    plt.style.use('ggplot')
    plt.show() 
    
    canvas = FigureCanvasTkAgg(fig, master = window3s)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH) 
    
    toolbar = NavigationToolbar2Tk(canvas, window3s)
    toolbar.update()
    canvas.get_tk_widget().pack(side=tk.BOTTOM, fill=tk.BOTH, expand=1,padx=3, pady=3)
    
def plot4_s():
    window4s = Toplevel(window)
    window4s.title("New Window")
    window4s.geometry("700x400")
    
    func5=df1_s.groupby(df1_s['Age range']).sum()
    
    fig=plt.Figure(figsize =(6,5), dpi = 100)
    ax4s = fig.add_subplot(111)
    ax4s.set_title('Breakdown of age ranges for stop and search for Tames Walley Police')
    
    func5.plot(kind='pie', y='All_Age_Case', subplots=True, shadow = True,startangle=90,figsize=(15,10), autopct='%1.1f%%', ax=ax4s)
    plt.style.use('ggplot')
    plt.show(); 
    
    canvas = FigureCanvasTkAgg(fig, master = window4s)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH) 
    
    toolbar = NavigationToolbar2Tk(canvas, window4s)
    toolbar.update()
    canvas.get_tk_widget().pack(side=tk.BOTTOM, fill=tk.BOTH, expand=1,padx=3, pady=3)
    
    
    




window = tk.Tk()
window.geometry("1080x640")
window.wm_title("Project")

pw = ttk.PanedWindow(window, orient = tk.HORIZONTAL)
pw.pack(fill = tk.BOTH, expand = True)

w2 = ttk.PanedWindow(pw, orient = tk.VERTICAL)

frame1 = ttk.Frame(pw, width = 720, height = 640, relief = tk.SUNKEN)
frame2 = ttk.Frame(pw, width = 360, height = 320, relief = tk.SUNKEN)
frame3 = ttk.Frame(pw, width = 360, height = 320, relief = tk.SUNKEN)

w2.add(frame2)
w2.add(frame3)

pw.add(w2)
pw.add(frame1)

top_label2= tk.Label(frame2,text="Covid cases Analysis",font=("Cambria",20,"bold"),bg="#2595f7",fg="#ffffff",pady=10,relief="raised",height = 1, width = 30)
top_label2.grid(row = 0, column = 0, padx = 5, pady = 5)

covid_button0 = tk.Button(frame2, command = graph1,height = 2, width = 80,
                      text = "Total number of cases reported each day over a given date range to All Areas",bg='#f0cdcb',fg="#0d95f5")
covid_button0.grid(row=1,column=0,padx = 5, pady = 5)

covid_button1 = tk.Button(frame2, command = plot,height = 1, width = 80,
                      text = "Total number of cases reported each day over a given date range in Oxfordshire",bg="#e0d48d",fg="#00628f")
covid_button1.grid(row=2,column=0,padx = 5, pady = 5)


covid_button2 = Button(frame2, command = plot1,height = 1, width = 80,
                      text = "Total number of cases reported each week over a given date range in Oxfordshire",bg="#e0d48d",fg="#00628f")
covid_button2.grid(row=3,column=0,padx = 5, pady = 5)

covid_button3 = Button(frame2, command = plot2,height = 1, width = 80,
                      text = "Total number of cases reported each month over from the start of reporting to end in Oxfordshire",bg="#e0d48d",fg="#00628f")
covid_button3.grid(row=4,column=0,padx = 5, pady = 5)

covid_button4 = Button(frame2, command = plot3,height = 1, width = 80,
                      text = "Areas with the highest number of cases on a given day",bg="#e0d48d",fg="#00628f")
covid_button4.grid(row=5,column=0,padx = 5, pady = 5)

covid_button5 = Button(frame2, command = plot4,height = 1, width = 80,
                      text = "Areas with the largest positive change of cases for a given 7 day period",bg="#e0d48d",fg="#00628f")
covid_button5.grid(row=6,column=0,padx = 5, pady = 5)

covid_button6 = Button(frame2, command = plot5,height = 1, width = 80,
                      text = "Comparison of two or more areas concerning their daily change over a given date range",bg="#e0d48d",fg="#00628f")
covid_button6.grid(row=7,column=0, padx = 5, pady = 5)

covid_button7 = Button(frame2, command = plot6,height = 1, width = 80,
                      text = "Comparison of two or more areas concerning their cumulative cases total over a given date range",bg="#e0d48d",fg="#00628f")
covid_button7.grid(row=8,column=0, padx = 5, pady = 5)


top_label2= Label(frame3,text="Stop and Search Tames Walley Police",font=("Cambria",20,"bold"),bg="#2595f7",fg="#ffffff",pady=10,relief="raised",height = 1, width = 30)
top_label2.grid(row=0,column=0,padx = 10, pady=10)

stop_button1 = tk.Button(frame3, command = graph1_s,height = 2, width = 80,
                      text = "All Age Range Stop and Search Tames Walley Police Each Month",bg='#f0cdcb',fg="#0d95f5")
stop_button1.grid(row=1,column=0,padx = 5, pady = 5)

stop_button2 = tk.Button(frame3, command = plot2_s, height = 1, width = 80,
                      text = "Teen Age Stop and Search Tames Walley Police each month",bg="#e0d48d",fg="#00628f")
stop_button2.grid(row=2,column=0,padx = 5, pady = 5)

stop_button3 = tk.Button(frame3, command = plot3_s, height = 1, width = 80,
                      text = "Teen Age Stop and Search Tames Walley Police in July",bg="#e0d48d",fg="#00628f")
stop_button3.grid(row=3,column=0,padx = 5, pady = 5)

stop_button4 = tk.Button(frame3, command = plot4_s, height = 1, width = 80,
                      text = "Breakdown of age ranges for stop and search for Tames Walley Police",bg="#e0d48d",fg="#00628f")
stop_button4.grid(row=4,column=0,padx = 5, pady = 5)

quit = tk.Button(frame3, text = "Close Window",command = window.destroy,height = 2, width = 15,bg="#f51818",fg="#080000")
quit.grid(row=12,column=0,padx = 25, pady = 25)





window.mainloop()
```
